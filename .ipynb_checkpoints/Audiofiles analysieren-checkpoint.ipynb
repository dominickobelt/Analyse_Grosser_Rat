Dieses Script wertet die Redezeiten der Grossrätinnen und Grossräte des Kantons Aargau aus

```python
#Imports

import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import re
import datetime
from datetime import datetime, timedelta
import locale
locale.setlocale(locale.LC_TIME, 'de_DE.UTF-8')
```

```python
#FUNKTIONEN

#Diese Funktion rechnet die Redezeit und die Anzahl Wortmeldungen zusammen und gibt die Werte zurück
def get_times(id_parl,offset,time_ges,anzahl_voten):
    url_votes = "https://ag.recapp.ch/viewer/api/search/?language_id=undefined&deputy_id="+str(id_parl)+"&limit=50&offset="+str(offset)
    r = requests.get(url_votes)
    try:
        votes_dict = json.loads(r.text)
    except:
        print("Die Zahl "+str(anzahl_voten)+" ist genau durch 50 teilbar, darum hat der Seitenaufruf nicht geklappt")
        return(1,time_ges,anzahl_voten)
    for sitzung in votes_dict["data"]:
        date=sitzung["sessionDate"]
        date = pd.to_datetime(date, format='%Y-%m-%d')

        if date >= startdatum and date <= enddatum:
            for object in sitzung["objects"]:
                for segment in object["segments"]:
                    time_ges+=segment["segmentDuration"]
                    anzahl_voten+=1
                    if anzahl_voten%50==0:
                        offset+=50
                        return(offset,time_ges,anzahl_voten)
                    
    return(1,time_ges,anzahl_voten)

def fraktion(partei):
    if partei == "SVP" or partei == "EDU":
        fraktion = "SVP/EDU"
    else:
        fraktion = partei
    return fraktion

#Variablen
url_ids = "https://ag.recapp.ch/viewer/api/speaker"
r = requests.get(url_ids)
response_dict = json.loads(r.text)
offset = 0
anzahl_voten = 0
time_ges = 0

#Hier definieren wir den Zeitraum, den wir untersuchen möchten
startdatum = pd.to_datetime('2021-01-01', format='%Y-%m-%d')
enddatum = pd.to_datetime('2021-12-31', format='%Y-%m-%d')
```

```python
#Alle Zeiten mit Hilfe der Funktion get_times zusammenzählen
for elem in response_dict:
    offset = 0
    time_ges = 0
    anzahl_voten = 0
    while offset%50 == 0:
        offset,time_ges,anzahl_voten = get_times(elem["id"],offset,time_ges,anzahl_voten)
        elem["time"] = time_ges
        elem["statements"] = anzahl_voten
    print("Habe für "+elem["deputylastname"]+" eine Zeit notiert: "+str(time_ges)+" und so viele Statements: "+str(anzahl_voten))
                                             
```

```python
#Ergebnisse in ein Dataframe packen
df_speaktime = pd.DataFrame(response_dict).sort_values(by=['time'], ascending=True)
```
